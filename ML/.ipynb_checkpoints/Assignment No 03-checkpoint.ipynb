```python
del df['Serial No.']
```

```python
df.head()
```

```python
x= df.iloc[:,:-1]
x.head()
```

```python
y=df['Chance of Admit ']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pre=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cnf=confusion_matrix(y_test, y_pre)
```

```python
sns.heatmap(cnf,annot=True)
```

```python
plt.figure(figsize=(14, 10))
tree.plot_tree(model, filled=True, fontsize=10)
```

```python
clf = tree.DecisionTreeClassifier(criterion='gini', max_leaf_nodes=5,max_depth=5).fit(X_train, y_train)
plt.figure(figsize=(14, 8))
tree.plot_tree(clf, filled=True, fontsize=16)
```

```python
ypre=clf.predict(X_test)
```

```python
cnf2=confusion_matrix(y_test, ypre)
```

```python
sns.heatmap(cnf2,annot=True)
```

```python
clf.score(X_test,y_test)
```
